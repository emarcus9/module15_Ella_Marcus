## 1Ô∏è‚É£ What Is a Nested Loop?

A **nested loop** is a loop inside another loop.

- The **outer loop** runs first
- For each iteration of the outer loop, the **inner loop runs completely**

Nested loops are commonly used when working with:
- grids
- tables
- combinations of values


```python
# Run this cell

for i in range(3):
    for j in range(2):
        print(i, j)
```

## 2Ô∏è‚É£ Execution Order (Very Important)

Think about nested loops as **repetition within repetition**.

The inner loop finishes **all of its iterations** before the outer loop moves on.

Trace the output below carefully.


```python
# Trace this output line by line

for row in range(2):
    print('row:', row)
    for col in range(3):
        print('   col:', col)
```

## 3Ô∏è‚É£ Nested Loops with Strings

Nested loops are often used to compare **each character** with every other character.


```python
word = 'cat'

for c1 in word:
    for c2 in word:
        print(c1, c2)
```

## 4Ô∏è‚É£ Nested Loops with Lists

Nested loops are useful when working with **lists of lists**.


```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

for row in matrix:
    for value in row:
        print(value)
```

## 5Ô∏è‚É£ Counting with Nested Loops

Nested loops are often used to **count patterns** or **matches**.


```python
count = 0

for i in range(3):
    for j in range(3):
        count += 1

count
```

## 6Ô∏è‚É£ When to Use Nested Loops

Use nested loops when:
- you need to compare **every item with every other item**
- you are processing rows and columns
- you are working with grids or tables

Avoid nested loops when a single loop will do.

## Question 2 ‚Äî Counting Pairs

Write a function that counts how many pairs `(i, j)` are produced
when using two nested loops with `range(n)` and `range(m)`.


```python
def count_pairs(n, m):
    total = 0
    for i in range(n):
        for j in range(m):
            total += 1
    print(f"{total} pairs are made")
# Example:
# count_pairs(2, 3)  # should count all (i, j) pairs
count_pairs(2, 3)
```

## Question 3 ‚Äî Character Pairs

Write a function that prints all **ordered pairs of characters**
from a string.


```python
def character_pairs(s):
    print("Ordered Pairs:")
    for letter in s:
        for i, x in enumerate(s):
            print(f"({letter}, {x})")

character_pairs('dog')
```

## Question 4 ‚Äî Grid Sum

Write a function that takes a list of lists of numbers
and returns the **sum of all values** using nested loops.


```python
def grid_sum(grid):
    total = 0
    for row in grid:
        for num in row:
            total += num
    return total

grid = [[1, 2], [3, 4]]
print(grid_sum(grid))
```
