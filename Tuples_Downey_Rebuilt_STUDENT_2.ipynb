## 1Ô∏è‚É£ Tuples

A **tuple** is an ordered, immutable collection of values.

- **Ordered**: elements have positions
- **Immutable**: elements cannot be changed after creation

Examples:
- `(3, 7, 10)`
- `(1, "apple", 3.14)`


```python
# Run this cell

t = (3, 7, 10)
t
```

## 2Ô∏è‚É£ Tuples vs Lists

| Feature | List | Tuple |
|-------|------|-------|
| Mutable | Yes | No |
| Syntax | `[ ]` | `( )` |
| Best for fixed data | No | Yes |

Once created, tuple values cannot be reassigned.


```python
# Lists are mutable
lst = [1, 2, 3]
lst[0] = 99
lst
```

```python
# Tuples are immutable
tup = (1, 2, 3)
tup[0] = 99  # Uncomment to observe the error
tup
```

## 4Ô∏è‚É£ Indexing and Slicing

Tuples support indexing and slicing, just like strings and lists.


```python
t = ('a', 'b', 'c', 'd')

# Try indexing and slicing
print(t[0])
print(t[-1])
print(t[1:3])
```

## 6Ô∏è‚É£ Tuple Assignment (Unpacking)

Tuples allow multiple assignment in one statement.


```python
# TODO: unpack the tuple into two variables
pair = (3, 7)

a, b = pair
print(a)
print(b)
```

## 7Ô∏è‚É£ Functions That Return Tuples

A function can return multiple values as a tuple.


```python
# TODO: write a function that returns the minimum and maximum

def min_max(numbers):
    minimum, maximum = min(numbers), max(numbers)
    return minimum, maximum

# Test
data = [4, 2, 9]
result = min_max(data)
print(result)
```
